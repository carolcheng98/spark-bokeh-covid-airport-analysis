```python
df_case = pd.read_csv("us_confirmed_cases.csv",header=[0,1])
```

```python
df_case
```

```python
df_case.columns = df_case.columns.map('_'.join)
```

```python
df_old = df_case.columns
tempList=[]
for col in df_case.columns:
        new_name = col.strip()
        new_name = "".join(new_name.split())
        new_name = new_name.replace('.','') # EDIT
        tempList.append(new_name) #Edit02
print(tempList)
```

```python
df_case.rename(
    columns={i:j for i,j in zip(df_old,tempList)}, inplace=True
)

df_case.rename(columns={'Province_State_Admin2':'Date'},inplace=True)
```

```python
df_case
```

```python
df_case['Date'] = pd.to_datetime(df_case['Date'], errors='coerce')
```

```python
df_case
```

```python
df_case.to_csv('us_confirmed_pcs_2.csv',index=False)
```
